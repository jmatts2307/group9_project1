```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = './House Price Prediction Dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Scatter plot: Bathrooms vs. Price
plt.figure(figsize=(10,6))
plt.scatter(df['Bathrooms'], df['Price'], alpha=0.5)
plt.title('Scatter plot: Bathrooms vs. Price')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Bar plot: Average Price per Bathroom count
avg_price_per_bathroom = df.groupby('Bathrooms')['Price'].mean().reset_index()
plt.figure(figsize=(10,6))
plt.bar(avg_price_per_bathroom['Bathrooms'], avg_price_per_bathroom['Price'])
plt.title('Bar plot: Average Price per Bathroom count')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
# Box plot: Price distribution per Bathroom count
plt.figure(figsize=(10,6))
df.boxplot(column='Price', by='Bathrooms', grid=False)
plt.title('Box plot: Price distribution per Bathroom count')
plt.suptitle('')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Count plot: Number of properties with each Bathroom count
bathroom_counts = df['Bathrooms'].value_counts().sort_index()
plt.figure(figsize=(10,6))
bathroom_counts.plot(kind='bar')
plt.title('Count plot: Number of properties per Bathroom count')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Number of Properties')
plt.grid(True)
plt.show()
```
